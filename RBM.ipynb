```python
from PIL import Image
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neural_network import BernoulliRBM

im = Image.open(r'C:\Users\Will\simpsons_dataset\simpsons_dataset\barney_gumble\pic_0000.jpg')
im = im.resize([100,100])
plt.imshow(im)
plt.show()
```

```python
images = []

i=1
plt.figure(figsize=(25,25))
for file in glob.glob(r'C:\Users\Will\simpsons_dataset\simpsons_dataset\barney_gumble\*.jpg'):
    im = Image.open(file)
    im = im.resize([50,50])
    im = im.convert('L')
    im = np.array(im)
    plt.subplot(11,10, i)
    plt.imshow(im)
    im = im.flatten() / 255
    images.append(im)
    i+=1
    
plt.tight_layout()

images *= 3
```

```python
print(len(images))
print(len(images[0]))
```

```python
rbm = BernoulliRBM(batch_size=100, n_components=106, random_state=0, verbose=1, n_iter=400, learning_rate=0.002)
rbm.fit(images)
```

```python
len(rbm.components_)
```

```python
comps = rbm.components_
comps[0].shape
```

```python
print(comps[0])
```

```python
plt.imshow(np.abs(comps[0].reshape((50,50))), cmap=plt.cm.gray_r,
               interpolation='nearest')
```

```python
plt.figure(figsize=(20, 15))
for i, comp in enumerate(rbm.components_):
    plt.subplot(11, 10, i + 1)
    plt.imshow(np.abs(comp.reshape((50,50))), cmap=plt.cm.gray_r,
               interpolation='nearest')
    plt.xticks(())
    plt.yticks(())
plt.suptitle('100 components extracted by RBM', fontsize=16)
plt.subplots_adjust(0.08, 0.02, 0.92, 0.85, 0.08, 0.23)
plt.show()
```
